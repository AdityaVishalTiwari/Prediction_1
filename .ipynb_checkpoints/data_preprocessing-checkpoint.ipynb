```python
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
df3['total_sqft'].value_counts()
```

```python
# We observe that 'total_sqft' contains string values in a different format
total_sqft_int = []
for str_val in df3['total_sqft']:
    try:
       total_sqft_int.append(float(str_val))
    except:
        try:
            temp=[]
            temp=str_val.split('-')
            total_sqft_int.append((float(temp[0])+float(temp[1]))/2)
        except:
            total_sqft_int.append(np.nan)
df4 = df3.reset_index(drop=True)
df5 = df4.join(pd.DataFrame({'total_sqft_int':total_sqft_int}))
df5.head()
```

```python
# Finding Outliers and removing
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

def diagnostic_plots(df, variable):
    plt.figure(figsize=(16, 4))
    plt.subplot(1, 3, 1)
    sns.histplot(df[variable], bins=30)  # Changed displot to histplot
    plt.title('Histogram')
    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.ylabel('Variable quantiles')
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')
    plt.tight_layout()  # Explicitly calling tight_layout
    plt.show()

num_var = ["bath", "balcony", "total_sqft_int", "bhk", "price"]
for var in num_var:
    print("*********** {} **********".format(var))
    diagnostic_plots(df7, var)
```

```python
# considering 1 bhk required min 350 sqft
df7[df7['total_sqft_int']/df7['bhk']<350].head()
```

```python
df8 = df7[~(df7['total_sqft_int']/df7['bhk']<350)]
df8.shape
```

```python
#creating a new feature price per square foot that will help to find outliers
df8 = df8.copy()  # Create a copy of the DataFrame
df8['price_per_sqft'] = df8['price'] * 100000 / df8['total_sqft_int']
```

```python
df8.price_per_sqft.describe()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df9 = remove_pps_outliers(df8)
df9.shape
```

```python
def plot_scatter_chart(df, location):
    bhk2 = df[(df.location==location) & (df.bhk==2)]
    bhk3 = df[(df.location==location) & (df.bhk==3)]
    plt.figure(figsize=(16,9))
    plt.scatter(bhk2.total_sqft_int,bhk2.price, color='Blue', label='2 bhk', s=50)
    plt.scatter(bhk3.total_sqft_int,bhk3.price, color='Red',label='3 bhk',s=50,marker="+")
    plt.xlabel("Total Squre Feet Area")
    plt.ylabel("price")
    plt.title(location)
    plt.legend()

plot_scatter_chart(df9, "Rajaji Nagar")
```

```python
plot_scatter_chart(df9, "Hebbal")
```

```python
# Removing bhk outliers
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')

df10 = remove_bhk_outliers(df9)
df10.shape
```

## Categorical Variable Encoding

```python
df13 = df11.drop(["size","total_sqft"], axis=1)
df13.head()
```

```python
df13.shape
```

```python
# Working on area_type feature
df13['area_type'].value_counts()
```

```python
df15 = df13.copy()
# applying One-Hot encoding on 'area_type' feature
for cat_var in ["Super built-up Area", "Built-up Area", "Plot Area"]:
    df15["area_type"+cat_var] = np.where(df15['area_type']==cat_var,1,0)
df15.shape
```

```python
df15.head()
```

```python
# Working with availability feature
df15["availability"].value_counts()
```

```python
df15["availability_Ready_To_Move"]=np.where(df15["availability"]=="Ready To Move",1,0)
df15.shape
```

```python
df15.tail()
```

```python
# Working on location feature
location_value_count = df15['location'].value_counts()
location_value_count
```

```python
location_value_count = df15['location'].value_counts()

# Select locations with 20 or more occurrences
location_gert_20 = location_value_count[location_value_count >= 20].index

# Create a copy of df15
df16 = df15.copy()

# Create a list to store new DataFrame objects
dfs_to_concat = []

# Iterate over selected locations and create binary columns
for cat_var in location_gert_20:
    new_column = np.where(df16['location'] == cat_var, 1, 0)
    dfs_to_concat.append(pd.DataFrame({f'location_{cat_var}': new_column}, index=df16.index))

# Concatenate all new DataFrame objects along axis=1
df16 = pd.concat([df16] + dfs_to_concat, axis=1)

# Print the shape of the resulting DataFrame
print(df16.shape)
```

```python
# Dropping categorical variable
df17 = df16.drop(["area_type","availability",'location'], axis =1)
df17.shape
```
