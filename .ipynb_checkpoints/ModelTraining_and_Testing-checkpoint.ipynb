```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

```python
df = pd.read_csv('processed_data.csv')
df.shape
```

```python
df.head()
```

```python
x = df.drop("price", axis=1)
y = df['price']
print('Shape of x = ', x.shape)
print('Shape of y = ', y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
print('Shape of x_train = ', x_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of x_test = ', x_test.shape)
print('Shape of y_test = ', y_test.shape)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
X_train= sc.transform(x_train)
X_test = sc.transform(x_test)
```

# Model Training

## Linear Regressing

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
lr = LinearRegression()
lr_lasso = Lasso()
lr_ridge = Ridge()
```

```python
def rmse(y_test, y_pred):
  return np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
lr.fit(x_train, y_train)
lr_score = lr.score(x_test, y_test)
lr_rmse = rmse(y_test, lr.predict(x_test))
lr_score, lr_rmse
```

```python
# Lasso 
lr_lasso.fit(X_train, y_train)
lr_lasso_score=lr_lasso.score(X_test, y_test) 
lr_lasso_rmse = rmse(y_test, lr_lasso.predict(X_test))
lr_lasso_score, lr_lasso_rmse
```

## Support Vector Machine

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train,y_train)
svr_score=svr.score(X_test,y_test)
svr_rmse = rmse(y_test, svr.predict(X_test))
svr_score, svr_rmse
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
rfr_score=rfr.score(X_test,y_test) 
rfr_rmse = rmse(y_test, rfr.predict(X_test))
rfr_score, rfr_rmse
```

## XGBoost

```python
import xgboost
xgb_reg = xgboost.XGBRegressor()
xgb_reg.fit(X_train,y_train)
xgb_reg_score=xgb_reg.score(X_test,y_test) 
xgb_reg_rmse = rmse(y_test, xgb_reg.predict(X_test))
xgb_reg_score, xgb_reg_rmse
```

```python
print(pd.DataFrame([{'Model': 'Linear Regression','Score':lr_score, "RMSE":lr_rmse},
              {'Model': 'Lasso','Score':lr_lasso_score, "RMSE":lr_lasso_rmse},
              {'Model': 'Support Vector Machine','Score':svr_score, "RMSE":svr_rmse},
              {'Model': 'Random Forest','Score':rfr_score, "RMSE":rfr_rmse},
              {'Model': 'XGBoost','Score':xgb_reg_score, "RMSE":xgb_reg_rmse}],
             columns=['Model','Score','RMSE']))
```

```python
from sklearn.model_selection import KFold,cross_val_score
cvs = cross_val_score(xgb_reg, X_train,y_train, cv = 10)
cvs, cvs.mean()
```

```python
cvs_rfr = cross_val_score(rfr, X_train,y_train, cv = 10)
cvs_rfr, cvs_rfr.mean()
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from xgboost.sklearn import XGBRegressor
```

```python
xgb1 = XGBRegressor()
parameters = {
    'learning_rate': [0.1, 0.03, 0.05, 0.07],
    'min_child_weight': [1, 3, 5],
    'max_depth': [4, 6, 8],
    'gamma': [0, 0.1, 0.001, 0.2],
    'subsample': [0.7, 1.0],  # Adjusted to include only valid values
    'colsample_bytree': [0.7, 1.0],  # Adjusted to include only valid values
    'objective': ['reg:squarederror'],  # Updated to 'reg:squarederror' for newer versions of XGBoost
    'n_estimators': [100, 300, 500]
}

# Perform grid search with adjusted parameters
xgb_grid = GridSearchCV(
    estimator=xgb1,
    param_grid=parameters,
    cv=2,
    n_jobs=-1,
    verbose=3  # Adjust verbosity for more detailed output
)
xgb_grid.fit(X_train, y_train)

# Print best score and best parameters
print(xgb_grid.best_score_)
print(xgb_grid.best_params_)
```

```python
xgb_tune = xgb_grid.estimator
xgb_tune.fit(X_train,y_train)
xgb_tune.score(X_test,y_test)
```

```python
cvs = cross_val_score(xgb_tune, X_train,y_train, cv = 10)
cvs, cvs.mean()
```

```python
rmse = np.sqrt(mean_squared_error(y_test, xgb_tune.predict(X_test)))
print(rmse)
```

```python
import warnings
warnings.filterwarnings('ignore', 'X does not have valid feature names')
def predict_house_price(model, bath, balcony, total_sqft_int, bhk, price_per_sqft, area_type, availability, location):
    xo = np.zeros(len(model.feature_importances_))  
    xo[0] = bath
    xo[1] = balcony
    xo[2] = total_sqft_int
    xo[3] = bhk
    xo[4] = price_per_sqft

    if availability == "Ready To Move":
        xo[8] = 1

    area_type_col = 'area_type' + area_type
    if area_type_col in model.feature_importances_:
        area_type_index = np.where(model.feature_importances_ == area_type_col)[0][0]
        xo[area_type_index] = 1

    location_col = 'location_' + location
    if location_col in model.feature_importances_:
        loc_index = np.where(model.feature_importances_ == location_col)[0][0]
        xo[loc_index] = 1

    # Scaling the input features
    xo = sc.transform([xo])
    
    return model.predict(xo)[0]
```

```python
predict_house_price(model=xgb_tune, bath=3,balcony=2,total_sqft_int=1672,bhk=3,price_per_sqft=8971.291866,area_type="Plot  Area",availability="Ready To Move",location="Devarabeesana Halli")
```

```python
predict_house_price(model=xgb_tune, bath=3,balcony=2,total_sqft_int=1750,bhk=3,price_per_sqft=8571.428571,area_type="Super built-up",availability="Ready To Move",location="Devarabeesana Halli")
```
